```python
import pandas
```

```python
df = pandas.read_csv('../donaldTrumpTweets', sep='\t', encoding='latin1')
```

```python
df.head()
```

```python
#Parse out not available tweets 
availableTweets = df.loc[df['Tweet'] != 'Not Available']
availableTweets = availableTweets['Tweet']
```

```python
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
import string

stops = set(stopwords.words("english"))
availableTweets = availableTweets.apply(lambda x: x.lower())
availableTweets = availableTweets.apply(lambda x: ' '.join([word for word in TweetTokenizer().tokenize(x) if 'http' not in word and 'https' not in word]))
availableTweets = availableTweets.apply(lambda x: ' '.join([word for word in TweetTokenizer().tokenize(x) if '/' not in word]))
availableTweets = availableTweets.apply(lambda x: ' '.join([word for word in TweetTokenizer().tokenize(x) if '\\x' not in word]))
availableTweets = availableTweets.apply(lambda x: ' '.join([word for word in TweetTokenizer().tokenize(x) if word not in stops and '@' not in word]))
availableTweets = availableTweets.apply(lambda x: ' '.join([word for word in TweetTokenizer().tokenize(x) if word not in stops and '#' not in word]))
```

```python
# Split into train and test
from sklearn.model_selection import train_test_split
train, test = train_test_split(availableTweets, test_size=0.95)
```

```python
pos = map(lambda s: nltk.pos_tag(TweetTokenizer().tokenize(s)), train)
pos = list(pos)
```

```python
adj = []
for t in pos:
    for w in t:
        if 'JJ' in w[1]:
            adj.append(w[0])
            
adj = set(adj)
len(adj)
```
